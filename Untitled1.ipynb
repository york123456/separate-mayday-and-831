```python
# -*- coding: utf-8 -*-
"""
Created on Sun Oct 30 20:34:52 2022

@author: b4100
"""
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from pydub import AudioSegment
from pydub.playback import play
import time
from multiprocessing import Process



wavFile = "./rrrr/y2mate.com - .mp3"

# Retrieve the data from the wav file
data, samplerate = sf.read(wavFile)

n = len(data)  # the length of the arrays contained in data
Fs = samplerate  # the sample rate

# Working with stereo audio, there are two channels in the audio data.
# Let's retrieve each channel seperately:
ch1 = np.array([data[i][0] for i in range(n)])  # channel 1
ch2 = np.array([data[i][1] for i in range(n)])  # channel 2

# x-axis and y-axis to plot the audio data
time_axis = np.linspace(0, n / Fs, n, endpoint=False)
sound_axis = ch1 #we only focus on the first channel here


plt.plot(time_axis, sound_axis)
plt.show()

sample_rate = 500  # 

samples = sound_axis[0:n:10000]
n=int(n/10000)
sp = np.fft.fft(samples)
 


freq = np.fft.fftfreq(samples.size, d = 1 / sample_rate)
amp = np.abs(sp)
amp = np.log(amp)
amp = np.abs(amp)

ax = plt.gca()
ax.stem(freq, amp / np.max(amp),linefmt='--')
ax.set_xlim([0, np.max(freq)])
plt.show()
amp=amp[::5]
amp=amp[0:100]

ampA=amp[0:25]

for j in range(25):
        ampA[j]=max([amp[j*4+0],amp[j*4+1],amp[j*4+2],amp[j*4+3]])       

saved_model_path = "mnist-2"
model = tf.keras.models.load_model(saved_model_path)

#model.summary()

a=np.zeros((1,25))
a[0]=ampA

results = model(tf.constant(a))

predictions = np.argmax(model.predict(a), axis=-1)
# get prediction result
print(wavFile)
print(predictions)
print('\n:')
if predictions==0:
    print("")
if predictions==1:
    print("")
    




```
